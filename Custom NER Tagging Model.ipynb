```python
import spacy
from spacy.tokens import DocBin
from tqdm import tqdm

nlp = spacy.blank('en') # load a new spacy model
db = DocBin() # create a DocBin object
```

```python
import json
data = open('Training_Data_1.json')
TRAIN_DATA = json.load(data)
```

```python
for text, annot in tqdm(TRAIN_DATA['annotations']):
    doc = nlp.make_doc(text)
    ents = []
    for start, end, label in annot["entities"]:
        span = doc.char_span(start, end, label=label, alignment_mode="contract")
        if span is None:
            print("Skipping entity")
        else:
            ents.append(span)
    doc.ents = ents
    db.add(doc)

db.to_disk("./Training_Data_1.spacy") # save the docbin object
```

```python
import spacy
from spacy.tokens import DocBin
from tqdm import tqdm
import json
import logging
from sklearn.model_selection import train_test_split
from spacy.training.example import Example
import os
from pathlib import Path
import numpy as np

# Set up logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler('ner_training.log'),
        logging.StreamHandler()
    ]
)

class NERTrainer:
    def __init__(self, lang='en', model_name=None):
        """Initialize NER trainer with specified language and optional base model."""
        self.nlp = spacy.load(model_name) if model_name else spacy.blank(lang)
        self.metrics = {}

    def preprocess_data(self, data_path):
        """Load and preprocess training data with validation."""
        try:
            with open(data_path, 'r', encoding='utf-8') as f:
                data = json.load(f)

            # Split data into train/val sets
            train_data, val_data = train_test_split(
                data['annotations'],
                test_size=0.2,
                random_state=42
            )

            logging.info(f"Loaded {len(train_data)} training and {len(val_data)} validation examples")
            return train_data, val_data

        except Exception as e:
            logging.error(f"Error loading data: {str(e)}")
            raise

    def create_docbin(self, data, output_path):
        """Create and save DocBin with error handling and statistics."""
        db = DocBin()
        skipped_entities = 0
        processed_entities = 0

        for text, annot in tqdm(data):
            doc = self.nlp.make_doc(text)
            ents = []

            # Clean and validate entity spans
            for start, end, label in annot["entities"]:
                if end <= len(text) and start >= 0 and start < end:
                    span = doc.char_span(
                        start, end,
                        label=label,
                        alignment_mode="contract"
                    )
                    if span:
                        ents.append(span)
                        processed_entities += 1
                    else:
                        skipped_entities += 1
                else:
                    skipped_entities += 1
                    logging.warning(f"Invalid entity boundaries: {start}:{end} in text of length {len(text)}")

            doc.ents = ents
            db.add(doc)

        db.to_disk(output_path)
        logging.info(f"Processed {processed_entities} entities, skipped {skipped_entities} invalid entities")
        return processed_entities, skipped_entities

    def evaluate_model(self, model, test_data):
        """Evaluate model performance on test data."""
        examples = []
        for text, annot in test_data:
            doc = self.nlp.make_doc(text)
            example = Example.from_dict(doc, annot)
            examples.append(example)

        scores = model.evaluate(examples)
        self.metrics = {
            "ents_p": scores["ents_p"],
            "ents_r": scores["ents_r"],
            "ents_f": scores["ents_f"]
        }
        logging.info(f"Model Evaluation Metrics: {self.metrics}")
        return self.metrics

    def train(self, config_path, output_dir, train_path, dev_path):
        """Train model with specified configuration."""
        try:
            os.system(f"python -m spacy train {config_path} "
                     f"--output {output_dir} "
                     f"--paths.train {train_path} "
                     f"--paths.dev {dev_path} "
                     f"--training.eval_frequency 100 "
                     f"--training.patience 5")

            logging.info("Model training completed successfully")

        except Exception as e:
            logging.error(f"Training failed: {str(e)}")
            raise

def main():
    # Initialize trainer
    trainer = NERTrainer(model_name="en_core_web_lg")

    # Process data
    train_data, val_data = trainer.preprocess_data("Training_Data_1.json")

    # Create DocBins
    train_processed, train_skipped = trainer.create_docbin(
        train_data,
        "./train.spacy"
    )
    val_processed, val_skipped = trainer.create_docbin(
        val_data,
        "./dev.spacy"
    )

    # Train model
    trainer.train(
        "config.cfg",
        "./output",
        "./train.spacy",
        "./dev.spacy"
    )

    # Evaluate model
    model = spacy.load("./output/model-best")
    metrics = trainer.evaluate_model(model, val_data)

    # Save metrics
    with open("./output/metrics.json", "w") as f:
        json.dump(metrics, f, indent=2)

if __name__ == "__main__":
    main()
```
